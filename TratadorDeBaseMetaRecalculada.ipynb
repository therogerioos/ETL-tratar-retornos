```python
# @title
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import pandas as pd
import os
import sys
from datetime import datetime

# Definir o nome do arquivo a ser tratado
# IMPORTANTE: Inserir o nome do arquivo
nome_do_arquivo = 'ArquivoBruto_ControleControle_24052024'

parametroData = nome_do_arquivo.split("_")[2]
parametroOperation = nome_do_arquivo.split("_")[1]

nome_do_arquivo_baseVendas = f'BaseVendas_{parametroOperation}_{parametroData}'

caminho_do_arquivo = f'/content/drive/MyDrive/TIMrotinasPython/RetornoBruto/{nome_do_arquivo}.csv'


try:

    tabelaBruta = pd.read_csv(caminho_do_arquivo, sep=';', low_memory=False, on_bad_lines='skip')
    print(f"O arquivo: {nome_do_arquivo}.csv foi importado com sucesso!")

    tabelaBruta = tabelaBruta.drop('info_extra', axis=1)
    tabelaBruta = tabelaBruta.drop('agenda', axis=1)
    tabelaBruta = tabelaBruta.drop('usuario_retorno_codigo', axis=1)
    tabelaBruta = tabelaBruta.drop('usuario_retorno_codigo_descricao', axis=1)
    tabelaBruta = tabelaBruta.drop('ddd_agendado', axis=1)
    tabelaBruta = tabelaBruta.drop('telefone_agendado', axis=1)
    tabelaBruta = tabelaBruta.drop('receptivo', axis=1)
    tabelaBruta = tabelaBruta.drop('fidelizado', axis=1)
    tabelaBruta = tabelaBruta.drop('complemento', axis=1)
    tabelaBruta = tabelaBruta.drop('acd_id', axis=1)
    tabelaBruta = tabelaBruta.drop('descricao', axis=1)
    tabelaBruta = tabelaBruta.drop('objeto_status_descricao', axis=1)
    tabelaBruta = tabelaBruta.drop('objeto_status', axis=1)
    tabelaBruta = tabelaBruta.drop('pessoa_codigo', axis=1)

    filtro = ['Pre-Venda', 'Venda']

    itens_filtrados = tabelaBruta[tabelaBruta['historico_tipo_descricao'].isin(filtro)]

    itens_filtrados['ddd'] = itens_filtrados['ddd'].astype(int).astype(str)
    itens_filtrados['telefone'] = itens_filtrados['telefone'].astype(int).astype(str)

    itens_filtrados['dddTelefone'] = itens_filtrados['ddd'] + itens_filtrados['telefone']

    contagem_por_tipo = itens_filtrados['historico_tipo_descricao'].value_counts()
    contagem_por_tipo = contagem_por_tipo.reset_index()
    contagem_por_tipo.columns = ['Tabulação', 'Contagem']

    print("================================================================================================")

    print(contagem_por_tipo)

    print("================================================================================================")

    # Diretório onde deseja salvar o arquivo
    diretorio_salvar = '/content/drive/MyDrive/TIMDescargaBaseMetaRecalculada'

    nomeDoArquivoFinal = f'BaseMetaRecalculada_{parametroOperation}_{parametroData}.csv'

    # Caminho completo para salvar o arquivo
    caminho_arquivo = os.path.join(diretorio_salvar, nomeDoArquivoFinal)

    # Salvar a tabela em formato CSV
    itens_filtrados.to_csv(caminho_arquivo, encoding='latin-1', index=False)

    print(f"Concluído processo e arquivo: {nomeDoArquivoFinal} salvo!!")
    print(f"Diretório de armazenamento do arquivo: {diretorio_salvar}")
    print("================================================================================================")


except FileNotFoundError as e:
    # Captura a exceção e imprime qual arquivo não foi encontrado
    print(f"Arquivo não encontrado: {e}")
    print("Devido ao erro acima, o algoritmo foi encerrado")


```
